```python
import pandas as pd
import numpy as py

path = "C:\python_src\pandas-data-analysis\part7\data\San-Francisco-Crime .csv"

df = pd.read_csv(path)
df.head()
```

```python
df.describe(include='object')
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split
X = df[['X','Y']]
y = df['Category']

X_train,X_test,y_train,y_test = train_test_split(X,y, stratify=y, test_size=0.3, random_state=42)
len(X_train), len(X_test), len(y_test), len(y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier()

tree.fit(X_train,y_train)
y_hat = tree.predict(X_test)
```

```python
from sklearn.metrics import classification_report

tree_report = classification_report(y_test,y_hat)
print(tree_report)
```

```python
from sklearn.model_selection import GridSearchCV

tree= DecisionTreeClassifier()
param_grid = {
    'criterion' : ['gini', 'entropy', 'log_loss'],
    'max_depth' : range(1,10)
}

grid_searchcv = GridSearchCV(tree, param_grid=param_grid, cv=3, scoring='f1_macro')
grid_searchcv.fit(X_train,y_train)
```

```python
tree_new = DecisionTreeClassifier(criterion='entropy',max_depth=6)

tree_new.fit(X_train,y_train)
y_hat_new = tree_new.predict(X_test)
```

```python
tree_new_report = classification_report(y_test,y_hat_new)
tree_new_report
```

```python
import numpy as np
print("y_test 분포:", np.unique(y_test, return_counts=True))
print("y_hat_new 분포:", np.unique(y_hat_new, return_counts=True))
```

```python
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=42)
X_resampled, y_resampled = ros.fit_resample(X_train, y_train)
```

```python
tree_balanced = DecisionTreeClassifier(
    criterion='entropy',
    max_depth=10,
    class_weight='balanced',
    random_state=42
)
tree_balanced.fit(X_train, y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(
    n_estimators=200,
    max_depth=15,
    class_weight='balanced_subsample',
    random_state=42
)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
print(classification_report(y_test, y_pred, zero_division=1))
```
